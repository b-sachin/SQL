```python
df = pd.read_sql_query("""select p.productCode, p.productName, p.MSRP, pl.productLine, pl.textDescription from products p, productlines pl where p.productLine = pl.productLine""",dbcon, parse_dates = True)
```

### approach 1) Read it in multiple data frame --> use merge to join

```python
df1 = pd.read_sql_query("""select productCode, productName, MSRP, productLine from products""", dbcon, parse_dates = True)
```

```python
df2 = pd.read_sql_query("""select productLine, textDescription from productlines;""", dbcon, parse_dates = True)
```

```python
new_df = df1.merge(df2, how='inner', on='productLine')
```

```python
new_df
```
